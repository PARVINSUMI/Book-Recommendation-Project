```python
from kneed import KneeLocator
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

# Generate example data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Compute WCSS for different k values
wcss = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Automatically detect the elbow
knee = KneeLocator(k_values, wcss, curve='convex', direction='decreasing')
optimal_k = knee.knee

print(f"Optimal number of clusters: {optimal_k}")

# Plot the elbow graph with the detected elbow
plt.figure(figsize=(8,5))
plt.plot(k_values, wcss, marker='o')
plt.vlines(optimal_k, plt.ylim()[0], plt.ylim()[1], linestyles='dashed', colors='red')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS (Inertia)')
plt.title('Elbow Method with Automatic Elbow Detection')
plt.xticks(k_values)
plt.grid(True)
plt.savefig(output_folder)
plt.show()
```

```python
def recommend_similar_books(books_index, df, n_recommendations=5):
    """
    Recommend similar books based on cluster labels.
    """
    if 'cluster' not in df.columns:
        raise ValueError("DataFrame must have a 'cluster' column.")

    cluster_label = df.loc[books_index, 'cluster']
    similar_books = df[df['cluster'] == cluster_label].drop(index=books_index)

    return similar_books.sample(n=min(n_recommendations, len(similar_books)), random_state=42)

```

```python
# Recommended book

books_index = 10  # index of the book you want recommendations for
recommended_books = recommend_similar_books(books_index, df)
print(recommended_books)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Make a copy of the dataframe to avoid setting with copy warning
df_processed = df.copy()

# Fill NaN values with empty strings
features = ['Author', 'Categories', 'Publisher']
for feature in features:
    df_processed[feature] = df_processed[feature].fillna('')

# Combine all the features into a single string
def combine_features(row):
    return row['Author'] + ' ' + row['Categories'] + ' ' + row['Publisher']
df_processed['combined_features'] = df_processed.apply(combine_features, axis=1)

# Initialize TfidfVectorizer and create the matrix
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df_processed['combined_features'])

# Calculate the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Create a Series to map book titles to their indices
book_indices = pd.Series(df_processed.index, index=df_processed['Title']).drop_duplicates()

# Function to get book recommendations
def get_recommendations(title, cosine_sim=cosine_sim, df=df_processed, indices=book_indices):
    try:
        idx = indices[title]
    except KeyError:
        return "Book title not found in the dataset. Please check the spelling or try another book."

    # Get the pairwise similarity scores of all books with that book
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the books based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 5 most similar books
    sim_scores = sim_scores[1:6]  # Exclude the book itself

    # Get the book indices
    book_indices_list = [i[0] for i in sim_scores]

    # Return the top 5 most similar books
    return df['Title'].iloc[book_indices_list]

# Get and print recommendations for the book "1984"
recommendations = get_recommendations('1984')
print("\nTop 5 recommendations for '1984':")
if isinstance(recommendations, str):
    print(recommendations)
else:
    print(recommendations.to_markdown(index=False, numalign="left", stralign="left"))
```
